5. ให้นักศึกษา อธิบายผลที่ได้จาก run code ด้านล่าง

```python
def visual_multi(i):
  nplots = 40
  fig = plt.figure(figsize=(8, 4))
  for j in range(nplots):
    plt.subplot(4, 10, j + 1)
    plt.imshow(X_train[i + j], cmap=plt.cm.gray_r)
    plt.title(y_train[i + j])
    plt.xticks([]); plt.yticks([])
  plt.show()
```

```python
visual_multi(0)
```

6. ให้ นักศึกษา plot 5 แถว แถวละ 8 รูป แสดงผลด้านล่าง

```python
def visual_multi_custom(i):
  nplots = 40
  fig = plt.figure(figsize=(10, 6))
  for j in range(nplots):
    plt.subplot(5, 8, j + 1)
    plt.imshow(X_train[i + j], cmap=plt.cm.gray_r)
    plt.title(y_train[i + j])
    plt.xticks([]); plt.yticks([])
  plt.show()
visual_multi_custom(0)
```
